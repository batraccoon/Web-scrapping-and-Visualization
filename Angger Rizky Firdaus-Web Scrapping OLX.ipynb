## __Web Scrapping__

```python
# Syntax untuk mengaktifkan webdriver yang menggunakan browser yang digunakan.
driver = webdriver.Edge()
```

```python
# syntax untuk menyimpan html yang terdapat pada website yang dibuka saat itu.
html = driver.page_source
```

```python
# syntax untuk menampilan html yang sudah disimpan pada variable html sebelumnya.
soup = BeautifulSoup(html,'html.parser')
```

fungsi beautiful soup dan prettify adalah untuk menampilkan html /kode abstrak diatas secara rapih.

```python
# Syntax untuk melakukan webscrapping.

# list kosong untuk menyimpan informasi yang dibutuhkan
list_Nama = []
list_kota = []
list_tahun = []
list_harga = []



olx_link = (f'https://www.olx.co.id/_c198?filter=make_eq_mobil-bekas-honda%2Colxautos_listing_eq_true')
driver.get(olx_link)


html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
# syntax untuk mencari class dan atributes dari html yang diinginkan(bagian yang diinginkan.)


elements = soup.find_all('div', {'class':"_2v8Tq"})

   
for el in elements:
    try:
        harga = el.find('span', {'data-aut-id':"itemPrice",
                                 'class' :'_1zgtX'
                                })
        list_harga.append(harga.get_text())
    except:
        list_harga.append(None)

    try:
        nama = el.find('div', {'class':"_2Gr10"})
        list_Nama.append(nama.get_text())
    except:
        list_Nama.append(None)

    try:
        tahun = el.find('div', {'class':"_21gnE"})
        list_tahun.append(tahun.get_text())
    except:
        list_tahun.append(None)

    try:
        kota = el.find('div', {'class':"_3VRSm"
                                })
        list_kota.append(kota.get_text())
    except:
        list_kota.append(None)


```

```python
print(len(list_harga))
print(len(list_Nama))
print(len(list_tahun))
print(len(list_kota))


```

```python
cleaned_data = []

for item in list_kota:
    cleaned_item = re.sub(r'\d{1,2} [A-Za-z]{3}', '', item)  # Menghapus format tanggal seperti "29 Jan"
    cleaned_item = re.sub(r'Hari ini', '', cleaned_item)  # Menghapus kata "Hari ini"
    cleaned_data.append(cleaned_item.strip())  # Memastikan tidak ada spasi yang tidak diinginkan
       
print(cleaned_data)
```

```python
years = []
kilometers = []

for item in list_tahun:
    year, km_range = item.split(' - ')
    years.append(year)
    kilometers.append(km_range)
```

```python
# Syntax untuk membuat dataframe/tabel untuk hasil webscraping website tokopedia tadi.
df = pd.DataFrame({
    'nama_mobil' : list_Nama,
    'harga_mobil' : list_harga,
    'tahun' : years,
    'kilometers': kilometers,
    'kota' : cleaned_data
})
```
